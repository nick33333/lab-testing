```python
'''
Ensure that consistent protein_ids are used so I can look em up in uniprot

!!!
Instead of searching for gene IDs, look at protein IDs which are typically from GenBank. This might
be a more direct way of finding good info. Also, these are probably only found in the rows
where feature == protein

- For aa sequence refseq: on uniprot use refseq protein_id to uniprotkb (protein_id good)
- Look up uniprot protein ids associated with S. Aureus MGEs, mark them to annotate test set
- Finish up synteny filter
'''

'''
Notes for implementation:

- Use list of gtf indices instead of list of genes since there are duplicate genes
'''

def build_gene_dict(idx: "int", gtfs: "list[dataframe]") -> "dict":
    '''
    For a given dataframe, return a dict of entires with duplicate protein_ids and their
    coordinates
    '''
    geneDF = gtfs[idx][gtfs[idx]["feature"] == "CDS"] # ordered gene df
    geneDict = {cds:[0, []] for cds in geneDF["protein_id"]} # dict where keys are protein_ids and vals are lists[occurrences, coordinates]
    for cds in geneDF.index:
        row = geneDF.loc[cds]
        geneDict[ row["protein_id"] ][0] += 1
        geneDict[ row["protein_id"] ][1].append(cds)
    return geneDict

def all_unique(geneList, geneDict):
    dupDict = dict()
    ans = len(geneList) == len(geneDict.keys())
    if ans == False:
        for thing in [ (cds, geneDict[cds]) for cds in geneDict.keys() if geneDict[cds][0] > 1]:
            print(thing)
    return ans

geneDicts = [build_gene_dict(idx, gtfs) for idx in range(11)]
```
