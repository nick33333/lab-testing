```python
def geneIntersect(query: "dataframe", ref: "dataframe") -> "dataframe":
    '''
    Function for querying gene locations to dermine annotated genes from found gene locations
    Gonna have to go query to many times on ref in terms of dataframes
     
    Each query entry has to check for an intersection using formula in https://stackoverflow.com/questions/3269434/whats-the-most-efficient-way-to-test-if-two-ranges-overlap
    for all entries in query
    
    Annotate query rows which correspond to a gene, otherwise make default entry something like MISSING.
    This will help in synteny calculations.
    
    query: start, end
    ref:   txstart, txend
    '''
    
    qStarts, qEnds = query["start"], query["end"] # series
    rStarts, rEnds = ref["txstart"], ref["txend"] # series
    
    for qS, qE in (qStarts, qEnds):
        for rS, rE in (rStarts, rEnds):
            if qS > rE:
                break
            
    
    
    return query["start"]

    
# def isIntersect(start: "int", end: "int", strand: "str", ref: "dataframe") -> "bool":
#     if strand != ref['strand']:
#         return False
#     else:
#         if ref.query("@ref.start <= @start and @ref.end >= @end"): # If query is between bounds
#             return
#         if ref.query("@ref.start <= @start and @ref.end <= @end"): # If query's end is less than upper bound
#             return True
#         if ref.query("@ref.start >= @start and @ref.end >= @end"): # If query's start is greater than lower bound
#             return True
#         if ref.query("@ref.start >= @start and @ref.end <= @end"): # If bounds are between query start and end
#             return True
#         return False
```
